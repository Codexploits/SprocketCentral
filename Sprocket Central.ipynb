```python
# Convert 'DOB' to datetime format with the correct format
df3['DOB'] = pd.to_datetime(df3['DOB'], format='%d/%m/%Y', errors='coerce')

# Calculate age as timedelta
df3['Age'] = datetime.now() - df3['DOB']

# Convert timedelta to years
df3['Age'] = df3['Age'] / pd.Timedelta(days=365.25)

# Round the 'Age' column to zero decimal places
df3['Age'] = df3['Age'].round(0)

# Display the final result
df3.head()
```

```python
from datetime import datetime

# Filter out rows with problematic values
Customer_demographics = result_df[result_df['DOB'].str.len() < 20]
Customer_demographics.head()
# Convert 'DOB' to datetime format with the correct format
Customer_demographics['DOB'] = pd.to_datetime(Customer_demographics['DOB'], format='%d/%m/%Y', errors='coerce')
# Calculate age as timedelta
Customer_demographics['Age'] = datetime.now() - Customer_demographics['DOB']

# Convert timedelta to years
Customer_demographics['Age'] = Customer_demographics['Age'] / pd.Timedelta(days=365.25)
# Round the 'Age' column to zero decimal places
Customer_demographics['Age'] = Customer_demographics['Age'].round(0)
# Display the final result
Customer_demographics.head()
```

1.5 Customer sales analysis
We begin by calculating the total value of overall items bought by each customer over the given period and add that column to the table.
This gives us the Customer lifetime value of each customer.
Customer Lifetime Value (CLV) is a crucial metric that quantifies the total value a customer is expected to bring to a business over the entire duration of their relationship. Understanding how CLV affects business growth is essential for strategic planning and long-term success. Here's an explanation of how CLV influences business growth:

Revenue Generation:

High CLV implies that a customer is likely to make repeat purchases and generate revenue over an extended period.
Businesses with customers exhibiting high CLV can rely on a consistent and predictable revenue stream, contributing to sustained growth.
Customer Retention:

CLV is often closely tied to customer loyalty and retention. A higher CLV suggests that customers are satisfied with the products or services, leading to long-term relationships.
Focusing on customer retention over acquiring new customers can be cost-effective and contribute to stable revenue growth.
Marketing Efficiency:


```python
# Calculate total sales for each customer
total_sales_per_customer = fact_table.groupby('customer_id')['list_price'].sum().reset_index()
total_sales_per_customer.head()
```

```python
# Merge the total sales back to the original DataFrame
Customers_value_table = pd.merge(Customer_demographics, total_sales_per_customer, on='customer_id', how='left')
Customers_value_table.head()
```

```python
# Rename the 'list_price' column to 'CLV'
Customers_value_table.rename(columns={'list_price': 'CLV'}, inplace=True)
```

```python
Customers_value_table.head()
```

Understanding CLV allows businesses to allocate marketing resources more efficiently. It helps identify the most valuable customer segments and tailor marketing efforts to retain and upsell to these segments.
Businesses can optimize their marketing spend by concentrating on activities that yield the highest CLV.
Cross-Selling and Upselling:

High CLV customers are more likely to be receptive to cross-selling and upselling efforts. They have already demonstrated a willingness to engage with the brand, making them prime targets for additional products or services.
Effective cross-selling and upselling contribute directly to increased revenue and business growth.
Referral and Word of Mouth:

Satisfied customers with high CLV are more likely to become brand advocates, referring friends and family to the business.
Word-of-mouth marketing generated by high CLV customers can lead to organic customer acquisition, supporting business expansion.
Investor Confidence:

```python
# Display distinct values in the 'wealth_segment' column
distinct_wealth_segments = Customers_value_table['wealth_segment'].unique()

print("Distinct values in the 'wealth_segment' column:")
print(distinct_wealth_segments)
```

```python
# Calculate total CLV according to wealth segment
total_clv_by_wealth = Customers_value_table.groupby('wealth_segment')['CLV'].sum().reset_index()
total_clv_by_wealth.head()
```

```python
total_clv_by_wealth.to_csv(r'C:\Users\User\Downloads\Wealth_segment.csv', index=False)

print("DataFrame exported to CSV successfully.")
```

```python
# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(total_clv_by_wealth['CLV'], labels=total_clv_by_wealth['wealth_segment'], autopct='%1.1f%%', startangle=90)
plt.title('Total CLV by Wealth Segment (Pie Chart)')
plt.show()
```

```python
# Plot Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(total_clv_by_wealth['wealth_segment'], total_clv_by_wealth['CLV'], color=['blue', 'orange', 'green'])
plt.xlabel('Wealth Segment')
plt.ylabel('Total CLV')
plt.title('Total CLV by Wealth Segment (Bar Chart)')
plt.show()
```
